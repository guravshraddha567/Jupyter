```python
# importing modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings("ignore")
```

```python
disney=pd.read_csv("/content/disney")
disney.head()    # first five records
```

```python
disney.tail()      #last five records
```

```python
#Checking number of rows and columns
disney.shape
```

```python
#infromation about the dataset
disney.info()
```

```python
#Checking Null Values
disney.isnull().sum()
```

```python
#Statistical info
disney.describe().T
```

```python
#Checking correlation
disney.corr().T
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(disney.corr(), annot=True, cmap='summer')
plt.show()
```

```python
#distplot for checking the total gross in India
sns.distplot(disney.total_gross, color='Blue')
plt.title("Total gross in India")
```

```python
sns.distplot(disney.inflation_adjusted_gross,color='Red')
plt.title("Inflattion adjusted gross in india")
```

```python
sns.distplot(disney.total_gross,color='Yellow')
plt.title("total gross in India",fontsize=15)
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='total_gross',y='inflation_adjusted_gross',palette='CMRmap',data=disney)
plt.title("The total gross per movies",fontsize=15)
plt.xticks(rotation=90)
```

```python
disney['movie_title'].tolist()
```

```python
plt.figure(figsize=(8,8))
labels = disney.index
plt.pie(x='total_gross',data=disney[:10],labels='movie_title',startangle=90,autopct='%.1f%%')
plt.title("Top 5 movies in India", fontsize = 24)
```

```python
plt.figure(figsize=(14,10))
sns.pointplot(x='movie_title', y="total_gross", data=disney, color='Red')
plt.title("PointPlot: The total gross per movies :", fontsize=15)
#plt.show()
plt.xticks(rotation=90)
```

```python
sns.barplot(x="movie_title", y="total_gross", data=disney[:10])
plt.title("Top 10 movies:", fontsize=15)
plt.xticks(rotation=90)
```

```python
sns.jointplot(x='total_gross', y="movie_title", data=disney)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(7,7))
labels=disney.index
plt.pie(x='total_gross', data=disney[:5], labels='movie_title', startangle=90, autopct='%.1f%%')
plt.title("total gross in Top 5 Sates in India")
plt.show()
```

```python
plt.figure(figsize=(7,7))
labels=disney.index
plt.pie(x='inflation_adjusted_gross', data=disney[:5], labels='movie_title', startangle=90, autopct='%.1f%%')
plt.title("inflation adjusted grossn in Top 5 movies ", fontsize=24)
plt.show()
```

```python
plt.figure(figsize=(7,7))
labels=disney.index
plt.pie(x='total_gross', data=disney[:5], labels='genre', startangle=90, autopct='%.1f%%')
plt.title("total gross as per genere", fontsize=24)
plt.show()
```
